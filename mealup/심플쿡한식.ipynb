```python
import urllib.request as req
import urllib.parse as pa
from bs4 import BeautifulSoup as bs
from urllib.parse import quote
import pandas as pd
```

```python
# url = 'http://sim-cook.com/product/list.html?cate_no=51'
url = 'http://sim-cook.com/product/list.html?cate_no=51&page=2'
ww = req.urlopen(url).read().decode('utf-8')
soup = bs(ww, 'html.parser')
```

## 이름 가격 사진

```python
img = soup.select('div > div.prdImg > a > img')
name = soup.select('li > div > strong.name')
price = soup.select('ul > li.xans-record- > span')
# price
상품이름 = []
상품사진 = []
상품가격 = []
```

```python
img = soup.select('div > div.prdImg > a > img')
name = soup.select('li > div > strong.name')
price = soup.select('ul > li.xans-record- > span')
```

```python
for i in img:
    # print(i['src'])
    상품사진.append(i['src'])
```

```python
for i in name:
    # print(i.text)
    상품이름.append(i.text[6:])
```

```python
ss = []
for i in price:
    ss.append(i.text)
for i in range(2,len(ss),5):
    상품가격.append(ss[i])
    # print(ss[i])
```

```python
print(len(상품이름))
print(len(상품사진))
print(len(상품가격))
```

```python
df = pd.DataFrame({'상품이름':상품이름, '상품가격':상품가격, '상품사진링크':상품사진})
```
